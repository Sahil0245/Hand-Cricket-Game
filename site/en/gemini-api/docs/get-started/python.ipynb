```
import numpy as np

# a) Create an array X with random entries between 0 and 1
# Shape (10,5,4,3) - This represents 10 RGB image samples of size 5x4
X = np.random.rand(10, 5, 4, 3)  # You can change (10, 5, 4, 3) as needed

# b) Slice out the 2nd sample (index 1) of X into an array Y
Y = X[1]  # Change 1 to get a different sample

# c) Set the 2 lowermost rows of the first sample (index 0) to -1
X[0, -2:, :, :] = -1  # Change `0` for a different sample and `-2:` for different rows

# d) Slice out the 5th sample (index 4) to a new array W, keeping every 2nd row
W = X[4, ::2, :, :]  # Change `4` for a different sample, `::2` for a different row step

# e) Copy out samples with indices 0, 1, 3 to a new array Z using a single assignment
Z = X[[0, 1, 3]]  # Modify `[0, 1, 3]` to copy different samples

# f) Copy out all samples whose sum is smaller than 40 into a new array M
M = X[np.sum(X, axis=(1, 2, 3)) < 40]  # Change `40` to modify the sum threshold

# Output the results for verification
print("Shape of X:", X.shape)
print("Shape of Y (2nd sample):", Y.shape)
print("Shape of W (5th sample, every 2nd row):", W.shape)
print("Shape of Z (selected indices 0,1,3):", Z.shape)
print("Shape of M (samples with sum < 40):", M.shape)
```

```
import numpy as np
import matplotlib.pyplot as plt

# a) Plot the function x^4 between 0 and 2 using precisely 200 support points
x = np.linspace(0, 2, 200)  # 200 evenly spaced points between 0 and 2
y = x**4  # Compute x^4

plt.figure(figsize=(6, 4))
plt.plot(x, y, label='$x^4$', color='b')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of $x^4$ from 0 to 2')
plt.legend()
plt.grid()
plt.show()

# b) Generate 10 random "images" of shape (19,19) with values between 0 and 1, and display them in a single window
images = np.random.rand(10, 19, 19)  # 10 images of shape (19,19)

fig, axes = plt.subplots(2, 5, figsize=(10, 5))  # 2 rows, 5 columns grid
for i, ax in enumerate(axes.flat):
    ax.imshow(images[i], cmap='gray')  # Display each image
    ax.axis('off')  # Hide axes
plt.suptitle("10 Random Images (19x19)")
plt.show()

# c) Generate 300 numbers distributed according to a uniform distribution between 0 and 2, and display their histogram
data = np.random.uniform(0, 2, 300)  # 300 values between 0 and 2

plt.figure(figsize=(6, 4))
plt.hist(data, bins=20, color='g', edgecolor='black', alpha=0.7)  # Histogram with 20 bins
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Uniformly Distributed Numbers (0 to 2)')
plt.grid()
plt.show()
```

```
import tensorflow as tf

# Define the function f(x) = sum(i * x_i^2) for i = 1 to 3
def f(x):
    indices = tf.range(1, tf.shape(x)[0] + 1, dtype=tf.float32)  # Generates [1, 2, 3] (Change 1 and 3 for different range)
    return tf.reduce_sum(indices * x**2)  # Computes sum(i * x_i^2)

# Define input vectors (Change values as needed)
a1 = tf.Variable([1.0, 2.0, 3.0])  # Example: (1,2,3)^T
a2 = tf.Variable([2.0, 0.0, 2.0])  # Example: (2,0,2)^T

# Compute function outputs
f_a1 = f(a1)
f_a2 = f(a2)

print("f(a1) =", f_a1.numpy())  # Output for a1
print("f(a2) =", f_a2.numpy())  # Output for a2

# Compute gradient ∇f at a1
with tf.GradientTape() as tape:
    tape.watch(a1)
    f_a1 = f(a1)

grad_f_a1 = tape.gradient(f_a1, a1)
print("∇f(a1) =", grad_f_a1.numpy())  # Gradient at a1
```
